```python
import shapely

l = open("9").read().split("\n")
c = [tuple(int(j) for j in i.split(",")) for i in l]
r = set()
for p in c:
    for q in c:
        r.add(
            (
                (max(p[0], q[0]), max(p[1], q[1])),
                (max(p[0], q[0]), min(p[1], q[1])),
                (min(p[0], q[0]), min(p[1], q[1])),
                (min(p[0], q[0]), max(p[1], q[1])),
            )
        )

popo = shapely.Polygon(c)

r = [list(x) for x in r]
p = [shapely.Polygon(x) for x in r]
print(max(x.area for x in p))
poly = max([x for x in p if x.within(popo)] ,key = lambda y: y.area)

poly.bounds
```
