```python
d = {"out": set()}
for line in open("11").read().split("\n"):
    a, b = line.split(": ")
    c = b.split(" ")
    d[a] = set(c)
    # print(line)
print("d", d)
indices = list(d.keys())
print(indices)

M =  np.zeros(dtype=np.int32, shape=(len(indices), len(indices)))
print(M)

for (k,vs) in d.items():
    i = indices.index(k)
    for v in vs:
        j = indices.index(v)
        print(i, j, v)
        M[i,j] = 1

you = indices.index("you")
out = indices.index("out")
print(you, out)

N = M.copy()

t = 0
for _ in range(100):
    t += N[you, out]
    N @= M
print("t", t)
```
